```python
def CompareMC2(Z, N, N2):
    
    if (Z!=6 and Z!=5):
        out_tree.Draw("s[0].Theta()>>o"+str(Z)+"(50,0,1)", "Z=="+str(Z))
    elif (Z==6):
        out_tree.Draw("s[0].Theta()>>o"+str(Z)+"(50,0,0.1)", "Z=="+str(Z))
    elif (Z==5):
        out_tree.Draw("s[0].Theta()>>o"+str(Z)+"(50,0,0.2)", "Z=="+str(Z))
    o = r.gDirectory.Get("o"+str(Z))
    o.Scale(1/N2)
    o.Sumw2()
    
    if (Z!=6 and Z!=5):
        print("tanf>>Z_"+str(Z)+"(100, 0, 1)", "s[0].W()-70 == "+str(Z)+" && k0>=4")
        tracks_V.Draw("s[0].Theta()>>Z_"+str(Z)+"(50, 0, 1)", "s[0].W()-70 == "+str(Z)+" && k0>=4")
    elif (Z == 6):
        tracks_V.Draw("s[0].Theta()>>Z_"+str(Z)+"(50, 0, 0.1)", "s[0].W()-70 >= "+str(Z)+" && s[0].W()-70<9 && k0>=4")
    elif (Z == 5):
        tracks_V.Draw("s[0].Theta()>>Z_"+str(Z)+"(50, 0, 0.2)", "s[0].W()-70 == "+str(Z)+" && s[0].W()-70<9 && k0>=4")
    h = r.gDirectory.Get("Z_"+str(Z))
    h.Scale(1/N)
    h.Sumw2(r.kFALSE)
    
    h.SetFillColor(0)
    h.SetLineColor(r.kBlue)
    h.SetLineWidth(2)
    h.SetLineStyle(2)

    o.SetFillColor(0)
    o.SetLineColor(r.kCyan-2)
    o.SetLineWidth(2)
    
    s = r.THStack("s", "s")
    s.Add(o)
    s.Add(h)
    
    s.SetTitle("MC_{rec} vs Data [k_{0}>=4, Z="+str(Z)+"];tan(#theta);Entries/N_{trk}")
    if (Z == 6):
        s.SetTitle("MC_{rec} vs Data [k_{0}>=4, Z>=6];tan(#theta);Entries/N_{trk}")
    #s.Draw("nostack")

    legend = r.TLegend(0.45,0.65,0.9,0.9)
    legend.SetTextFont(0)
    legend.SetTextSize(1.)
    if (Z != 6):
        legend.AddEntry(h, "#splitline{Z="+str(Z)+" (MC_{rec}), Entries = "+str(round(h.GetEntries(),0))+" ("+str(round(100*h.GetEntries()/N, 1) ) + " %)}{Mean: "+ str(round(h.GetMean(), 2)) + ", RMS = "+str(round(h.GetRMS(),2))+"}")
        legend.AddEntry(o, "#splitline{Z="+str(Z)+ " (Data), Entries = "+str(round(o.GetEntries(),0))+" ("+str(round(100*o.GetEntries()/N2, 1) ) + " %)}{Mean: "+ str(round(o.GetMean(), 2)) + ", RMS = "+str(round(o.GetRMS(),2))+"}")
    else:
        legend.AddEntry(h, "#splitline{Z>="+str(Z)+" (MC_{rec}), Entries = "+str(round(h.GetEntries(),0))+" ("+str(round(100*h.GetEntries()/N, 1) ) + " %)}{Mean: "+ str(round(h.GetMean(), 2)) + ", RMS = "+str(round(h.GetRMS(),2))+"}")
        legend.AddEntry(o, "#splitline{Z>="+str(Z)+ " (Data), Entries = "+str(round(o.GetEntries(),0))+" ("+str(round(100*o.GetEntries()/N2, 1) ) + " %)}{Mean: "+ str(round(o.GetMean(), 2)) + ", RMS = "+str(round(o.GetRMS(),2))+"}")
   
    return s, legend
    
```
